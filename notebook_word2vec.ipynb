```python
train.title[12]
```

```python
def tokenizador(texto:str):

  texto = texto.lower()
  lista_alfa_num = []

  for i in nltk.word_tokenize(texto):
    if i in punctuation:
      continue
    else:
      lista_alfa_num.append(i)

  return lista_alfa_num
```

```python
test_text = 'Texto Exemplo, 1234.'
tokenizador(test_text)
```

```python
def combinacao_vetorial_por_soma(palavras_numeros:list):
  
  vetor_resultante = np.zeros(300)
  
  for i in palavras_numeros:
    try:
      vetor_resultante += model.get_vector(i)
    except KeyError:
      if i.isnumeric():
        i = '0' * len(i)
      else:
        vetor_resultante += model.get_vector('unknow')
  return vetor_resultante
```

```python
palavras_num = tokenizador('texto Mikeias')
vetor_texto = combinacao_vetorial_por_soma(palavras_num)

print(f'{len(vetor_texto)}\n\n{vetor_texto}')
```

# Vetorização de DataFrame

```python
def matriz_vetores(textos):
  x= len(textos)
  y = 300
  matrix = np.zeros((x,y))

  for i in range(x):
    palavras_num = tokenizador(textos.iloc[i])
    matrix[i] = combinacao_vetorial_por_soma(palavras_num)

  return matrix

matrix_vetores_train = matriz_vetores(train.title)
matrix_vetores_test = matriz_vetores(test.title)
```

```python
print(matrix_vetores_train.shape)
print(matrix_vetores_test.shape)
```

# Treinamento de modelo

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_reg_log = LogisticRegression(max_iter=250)
model_reg_log.fit(matrix_vetores_train, train.category)
```

```python
model_reg_log.score(matrix_vetores_test, test.category)
```

```python
model_reg_log.score(matrix_vetores_test, test.category)
```

```python
y_true = test.category
y_pred = model_reg_log.predict(matrix_vetores_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true, y_pred))
```
